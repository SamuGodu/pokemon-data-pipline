```python
import json

with open("../data/raw/pokemon_1.json", "r") as file:
    data = json.load(file)
```

```python
print(data.keys())
```

```python
for key in data.keys():
    print(key)
```

```python
print(json.dumps(data["abilities"], indent=4))
```

```python
for item in data["abilities"]:

    slot = item["slot"]
    ability_name = item["ability"]["name"]
    is_hidden = item["is_hidden"]

    print(slot, ability_name, is_hidden)
```

```python
print(json.dumps(data["types"], indent=4))
```

```python
for item in data["types"]:

    slot = item["slot"]
    type_name = item["type"]["name"]
    type_url = item['type']['url']
    type_id = int(type_url.split('/')[-2])

    print(slot, type_id, type_name)
```
